```python
import random
import re

class RuleBot:
    # Responses
    negative_res = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")

    random_questions = (
        "How are you today?",
        "What have you been up to?",
        "Do you have any plans for the weekend?",
        "What's your favorite hobby?",
        "Have you seen any good movies lately?"
    )

    def __init__(self):
        self.alienbabble = {
            'describe_planet_intent': r'.*\b(your planet|earth)\b.*',
            'answer_why_intent': r'.*\bwhy\b.*',
            'about_klu': r'.*\b(kl university|klu)\b.*'
        }

    def greet(self):
        self.name = input("What is your name?\n")
        will_help = input(f"Hi {self.name}, I am Bot. Will you help me learn about your planet?\n").lower()
        if will_help in self.negative_res:
            print("Have a nice Earth day! Goodbye!")
            return
        self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply.lower() == command:
                print("Have a nice day! Goodbye!")
                return True
        return False

    def chat(self):
        reply = input(random.choice(self.random_questions) + "\n").lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply) + "\n").lower()

    def match_reply(self, reply):
        for intent, regex_pattern in self.alienbabble.items():
            found_match = re.match(regex_pattern, reply)
            if found_match:
                if intent == 'describe_planet_intent':
                    return self.describe_planet_intent()
                elif intent == 'answer_why_intent':
                    return self.answer_why_intent()
                elif intent == 'about_klu':
                    return self.about_klu()
        return self.no_match_intent()

    def describe_planet_intent(self):
        responses = (
            "Earth seems like a fascinating place. What do you like most about it?",
            "I've heard a lot about Earth. Can you tell me more about your experiences here?"
        )
        return random.choice(responses)

    def answer_why_intent(self):
        responses = (
            "I'm here to learn more about humans. What can you tell me?",
            "I want to understand your culture better. Can you help me?"
        )
        return random.choice(responses)

    def about_klu(self):
        responses = (
            "KL University is a reputed educational institution. Have you heard of it?",
            "KL University offers a variety of programs and courses. What do you think about higher education?",
            "At KL University, students are encouraged to grow academically and personally. Are you interested in learning more?"
        )
        return random.choice(responses)

    def no_match_intent(self):
        responses = (
            "That's quite interesting. How do you spend your free time?",
            "What else can you share about that?",
            "I'm curious, what are your thoughts on this?",
            "Tell me more about your day.",
            "What do you enjoy doing?",
            "How do you usually spend your weekends?"
        )
        return random.choice(responses)

bot = RuleBot()
bot.greet()
```
